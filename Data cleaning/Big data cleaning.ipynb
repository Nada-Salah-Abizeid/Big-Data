# Fixing is_retweet column

```python
true_count = 0
for value in df['is_retweet']:
    if value == True :
        true_count += 1
print(true_count)
```

```python
df['is_retweet'] = df['retweets'].apply(lambda x: True if x > 0 else False)
```

```python
true_count = 0
for value in df['is_retweet']:
    if value == True :
        true_count += 1
print(true_count)
```

# Removing rows with missing cells

```python
df['Sentiment'].fillna(0, inplace=True)
```

```python
num_missing_values = df.isnull().any(axis=1).sum()
num_missing_values
```

```python
df.dropna()
```

# Duplicated rows

```python
dup_count = df.duplicated().sum()
dup_count
```

```python
df.drop_duplicates(inplace=True)
```

# Applying sentiment analysis

```python
import textblob
from textblob import TextBlob
```

```python
df['Sentiment'] = df['text'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
df.head()
```

```python
df['Sentiment'] = df['Sentiment'].apply(lambda x: 'Positive' if x > 0 else 'Neutral' if x == 0 else 'Negative')
```
