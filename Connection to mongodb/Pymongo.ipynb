```python
import pymongo
```

```python
import pandas as pd
```

```python
#Connecting to mongodb
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
db = myclient["mongodb"]
```

# Uploading users collection

```python
df = pd.read_csv('users.csv')
df
```

```python
df['id'] = [i for i in range(1, 178490)]
df
```

```python
df.rename(columns={'id': '_id'}, inplace=True)
```

```python
df.head()
```

```python
collectionUser = db["users"]
```

```python
data_dict = df.to_dict(orient= 'records') 
```

```python
collectionUser.insert_many(data_dict)
```
